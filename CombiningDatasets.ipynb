```python
import pandas as pd

df = pd.read_csv("deepseek_vs_chatgpt.csv")
df.columns = df.columns.str.strip()
```

```python
#Concatenation
# Split the DataFrame into two parts and concatenate
df1 = df.iloc[:100]
df2 = df.iloc[100:200]

df_concat = pd.concat([df1, df2], axis=0)
print("Concatenated DataFrame shape:", df_concat.shape)
```

```python
#Merge
feedback_df = pd.DataFrame({
    "User_ID": df["User_ID"].head(10),
    "Feedback": ["Excellent", "Good", "Average", "Poor", "Great", "Bad", "Average", "Excellent", "Fair", "Good"]
})

merged_df = pd.merge(df, feedback_df, on="User_ID", how="left")
print(merged_df[["User_ID", "AI_Platform", "Feedback"]].head())
```

```python
# Right Join Example
right_joined = pd.merge(df, feedback_df, on="User_ID", how="right")
print("Right join result:\n", right_joined)

# Outer Join Example
outer_joined = pd.merge(df, feedback_df, on="User_ID", how="outer")
print("Outer join result:\n", outer_joined)
```

```python
#Aggregation and Grouping
agg_result = df.groupby("AI_Platform")["Response_Accuracy"].mean()
print("Average accuracy per platform:\n", agg_result)

agg_multiple = df.groupby("AI_Platform").agg({
    "User_Experience_Score": "mean",
    "Response_Accuracy": ["mean", "max"]
})
print("Grouped stats:\n", agg_multiple)
```

```python
#Pivot Table
pivot = df.pivot_table(
    values="User_Experience_Score",
    index="AI_Platform",
    columns="Language",
    aggfunc="mean"
)
print("Pivot Table:\n", pivot)
```
